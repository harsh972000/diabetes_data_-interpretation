```python
# here is class which describe all the above code in just three line after impletation it is describing the whole data


class meraplot:
    import seaborn as sns
    import matplotlib.pyplot as plt
    import pandas as pd
    from ipywidgets import interactive, interact
    
    
    def __init__(self, data):
        self.data = data
        self.columns = data.columns
        
        
        
        
    def make_hist(self, wid, leng, columns = 2):
        fil = self.data.select_dtypes('int' or 'float').columns
        print(fil)
        es = len(fil)
        row = (es/columns)
        rows = int(np.ceil(row))
        plt.figure(figsize = (wid, leng))
        for e,i in enumerate(fil):
            plt.subplot(rows,columns, e+1)
            sns.histplot(x= self.data[i])
       
    def make_object(self, wid, leng, columns = 2):
        fil = self.data.select_dtypes('object').columns
        print(fil)
        es = len(fil)
        row = (es/columns)
        rows = int(np.ceil(row))
        plt.figure(figsize = (wid, leng))
        for e,i in enumerate(fil):
            plt.subplot(rows,columns, e+1)
            sns.countplot(x= self.data[i])
            plt.xticks(rotation = 45)
            
            
    def relation_table(self, index, col):
        print(pd.crosstab(self.data[index], self.data[col]))
        
    def relation_bar(self, index, col):
        pd.crosstab(self.data[index], self.data[col]).plot(kind = 'bar')

        
        
        
    
            
    
            
        
```
